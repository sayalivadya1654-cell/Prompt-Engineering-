- ChatGroq creates the AI model.
- ChatGroq is LangChain's wrapper around Groq LLMs.
- Specifies which LLM to use like llama-3.3-70b-versatile.
- load_dotenv reads the .env file.
-loads GROQ_API_KEY into the environment.
- get_env Retrieves the API key and returns api key(gsk_xxxxxxxxx).
- llm Creates an authenticated connection to the Groq model.
- model: Selects the AI model.
- api_key: Authenticates your request.

```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv

load_dotenv(override=True)

llm = ChatGroq(model="llama-3.3-70b-versatile")

response = llm.invoke("What is Prompt Engineering?")
print(response.content)
```

- Summary
 | Concept                        | Description                                                                                  |
| ------------------------------ | -------------------------------------------------------------------------------------------- |
| **Groq**                       | A platform that provides fast inference for Large Language Models (LLMs).                    |
| **LLM (Large Language Model)** | An AI model trained to understand and generate human language. Examples: Llama, GPT, Gemini. |
| **LangChain**                  | A framework that simplifies building applications with LLMs.                                 |
| **ChatGroq**                   | LangChain's interface for connecting to Groq-hosted models.                                  |
| **Prompt**                     | The input or instruction you give to an AI model.                                            |
| **Prompt Engineering**         | The practice of designing prompts to produce accurate, relevant, and useful AI responses.    |
| **API Key**                    | A secret key used to authenticate your application with the Groq API.                        |
| **`.env` file**                | A configuration file used to securely store environment variables such as API keys.          |
| **`invoke()`**                 | A LangChain method that sends a prompt to the model and returns the generated response.      |
| **`response.content`**         | The text generated by the AI model.                                                          |
